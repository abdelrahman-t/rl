```python
from functional import seq
from operator import itemgetter
from functools import partial
from parse import parse
import re

import matplotlib.pyplot as plt
```

```python
pattern = re.compile(r'\d*/\d*:.*', flags=re.MULTILINE)
```

```python
log = open('log.txt', 'r').read()
matches = seq(re.findall(pattern, log))
```

```python
template = '{_}episode: {episode}, duration: {duration}, episode steps: {steps},{_}mean reward: {reward} {_}loss: {loss},{_}'
```

```python
parsed = matches.map(partial(parse, template))
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import show, output_file
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

%matplotlib inline
```

```python
output_notebook()

x = parsed.map(itemgetter('episode')).to_list()
y = top=parsed.map(itemgetter('loss')).to_list()
p = figure(x_range=x, plot_height=400, plot_width=950)
p.vbar(x=x, top=y, width=0.45)

show(p)
```
